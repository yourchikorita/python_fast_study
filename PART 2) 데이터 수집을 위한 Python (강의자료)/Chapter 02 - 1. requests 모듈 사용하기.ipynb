#### requests 모듈
1. http request/response를 위한 모듈
2. HTTP method를 메소드 명으로 사용하여 request 요청 예) get, post

```python
import requests
url = 'https://news.v.daum.net/v/20190728165812603'
```

#### get 요청하기
1. http get 요청하기
2. query parameter 이용하여 데이터 전달하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'
resp = requests.get(url)
resp.text
```

#### post 요청하기
1. http post 요청하기
2. post data 이용하여 데이터 전달하기

```python
url = 'http://www.kangcom.com/member/member_check.asp'
data={
    'id':'macmath22',
    'pwd':'test'
}
resp = requests.post(url,data=data)
resp.text
```

#### HTTP header 데이터 이용하기
1. header 데이터 구성하기
2. header 데이터 전달하기

```python
url = 'http://www.kangcom.com/member/member_check.asp'
headers={
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.100 Safari/537.36'
}
resp = requests.get(url,headers)
resp.text
```

#### HTTP response 처리하기
1. response 객체의 이해
2. status_code 확인하기
3. text 속성 확인하기

```python
import requests
url = 'https://news.v.daum.net/v/20190728165812603'

resp = requests.get(url)
if resp.status_code == 200 :
      resp.text
else:
    print('error')        
```
